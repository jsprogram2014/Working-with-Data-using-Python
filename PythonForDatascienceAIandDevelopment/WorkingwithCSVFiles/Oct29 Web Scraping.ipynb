```python
import requests
from bs4 import BeautifulSoup
    
## Iam going to do my First Web Scrape today
## Will be modifying original coursera code due to the enforcement of robot rules by wikipedia

UA = "FirstWikiScraper/0.1 (me@example.com)"  # put your contact here
    
url = "https://en.wikipedia.org/wiki/Lionel_Messi"
    
## sending a http request to wiki web server which will send a html pages as a response
myrequest=requests.get(url, headers={"User-Agent": UA, "Accept-Encoding": "gzip"}, timeout=20)
    
## since response which is a html pages will be stored as a text i am accessing that to parse it
htmlcont=myrequest.text
    
## now i am going to parse it using BeautifulSoup module
    
## going to create a BeautifulSoup object which will hold the parsed content
    
soup=BeautifulSoup(htmlcont,'html.parser')

# print(htmlcont)

## going to access specific elements content since its parsed by BeautifulSoup
## following line of code will search for all <a> tag and extract their details and store all of them as dicts in the list

links=soup.find_all('a')


for link in links:
    print(link.text)
    
```
