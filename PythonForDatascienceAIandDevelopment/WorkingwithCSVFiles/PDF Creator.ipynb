```python
# First, you will need to install the library: pip install fpdf

from fpdf import FPDF

class PDF(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 15)
        self.cell(0, 10, 'HIS 277: The First Crusade - Study Notes', 0, 1, 'C')
        self.ln(10)

    def chapter_title(self, title):
        self.set_font('Arial', 'B', 12)
        self.set_fill_color(200, 220, 255)
        self.cell(0, 6, title, 0, 1, 'L', 1)
        self.ln(4)

    def chapter_body(self, body):
        self.set_font('Arial', '', 11)
        self.multi_cell(0, 5, body)
        self.ln()

pdf = PDF()
pdf.add_page()

# Content Dictionary
notes = {
    "1. Geopolitical Context (c. 1000 AD)": 
    "EUROPE: Fragmented into various kingdoms (France, HRE, England) and Slavic/Baltic states.\n"
    "ISLAMIC WORLD: Divided between Sunni Abbasids (under Buyid control) and Shi'ite Fatimids (Egypt/North Africa).\n"
    "BYZANTINE EMPIRE: In decline, facing territory loss in Anatolia.",

    "2. The Catalyst: The Seljuk Turks":
    "EXPANSION: By 1100, Seljuks controlled land from the Aral Sea to the Persian Gulf and Anatolia.\n"
    "KEY CONFLICT: The Battle of Manzikert (1071) was a major Byzantine defeat. The Turks occupied Anatolia, threatening Constantinople.",

    "3. The Call to Arms":
    "EMPEROR ALEXIUS COMNENUS (r. 1081-1118): Byzantine Emperor who asked the West for help.\n"
    "POPE URBAN II (r. 1088-99): Former Cluny monk who responded.\n"
    "COUNCIL OF PIACENZA (March 1095): Urban allowed Alexius's envoys to speak.\n"
    "COUNCIL OF CLERMONT (Nov 1095): Urban preached the sermon calling for the Crusade.",

    "4. Motivations for the Crusade":
    "1. PRESTIGE: Enhanced the power of the Papacy.\n"
    "2. AID: Concern for the Byzantine Empire.\n"
    "3. RELIGION: 'Liberation' of Jerusalem.\n"
    "4. PEACE: Exporting violence to solve private warfare in Europe.",

    "5. The Waves of the Crusade":
    "A. PEASANTS' CRUSADE (First Wave)\n"
    "- Led by Peter the Hermit.\n"
    "- Infamous for the massacre of 800 Jews in Worms, Germany (May 18, 1096).\n\n"
    "B. PRINCES' CRUSADE (Military Elite)\n"
    "- Leaders: Raymond of Toulouse, Robert of Flanders, Godfrey of Bouillon, Bohemond of Taranto.\n"
    "- Routes: Converged on Constantinople via land and sea.",

    "6. Outcomes":
    "VICTORIES: Captured Nicaea (1097), Edessa & Antioch (1098), Jerusalem (1099).\n"
    "CRUSADER STATES ESTABLISHED:\n"
    "1. County of Edessa\n"
    "2. Principality of Antioch\n"
    "3. County of Tripoli\n"
    "4. Kingdom of Jerusalem",

    "7. Long-Term Context":
    "SUBSEQUENT CRUSADES: Continued through 1204 (4th Crusade sacked Constantinople).\n"
    "ECONOMY: Increased trade between Europe (Venice, Genoa) and the East by the 1200s.\n"
    "MILITARY ORDERS: Rise of groups like the Knights Templar."
}

# Generate PDF content
for title, body in notes.items():
    pdf.chapter_title(title)
    pdf.chapter_body(body)

# Save the file
pdf.output("HIS277_First_Crusade_Notes.pdf")

print("PDF generated successfully: HIS277_First_Crusade_Notes.pdf")
```
