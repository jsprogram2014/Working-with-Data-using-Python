```python
from bs4 import BeautifulSoup

html="<!DOCTYPE html><html><head><title>Page Title</title></head><body><h3><b id='boldest'>Lebron James</b></h3><p> Salary: $ 92,000,000 </p><h3> Stephen Curry</h3><p> Salary: $85,000, 000 </p><h3> Kevin Durant </h3><p> Salary: $73,200, 000</p></body></html>"

soup=BeautifulSoup(html,'html.parser')
# print(soup.contents)
print('######Raw Content')
print(soup.prettify())
print('')

#printing contents of the webpage
print('########Accessing the Elements')
print(type(soup.title))
print(soup.title)
print(soup.h3)
print('')

# accessing the childs
print('######Accessing the Childs')
print(soup.h3.b)
print(soup.h3.b.text)
print('')

# accessing the parent
print('########Accessing the Parents')
print(soup.h3.b.parent)
print('')

# accessing the siblings
print('########Accessing the Siblings')
print(soup.h3.next_sibling)
print(soup.p.next_sibling)
print(soup.p.previous_sibling)
print('')

# accessing the attributes of a tag
print('##########Accessing the Attributes')
print(soup.h3.b['id'])
print(soup.h3.b.attrs)
print(soup.h3.b.get('id'))
print('')

# Navigable String
print('####### Navigable String')
print(soup.h3.string)
print(soup.h3.b.string)
print(type(soup.h3.string))
unicode_string=str(soup.h3.string)
print(unicode_string)
print(type(unicode_string))
```
