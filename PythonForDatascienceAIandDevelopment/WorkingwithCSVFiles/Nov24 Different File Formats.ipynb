```python
## working with xml file

import pandas as pd
import xml.etree.ElementTree as etree

tree=etree.parse('samplexml.xml')
root=tree.getroot()

# coloums=['Name','Phone Number','Birthday']
datalist=[]

for node in root:
    name=node.find('name').text
    phonenumber=node.find('phonenumber').text
    birthday=node.find('birthday').text

    datalist.append({'Name':name, 'Phone Number':phonenumber, 'Birthday':birthday})

df=pd.DataFrame(datalist)
print(df)
    
```
