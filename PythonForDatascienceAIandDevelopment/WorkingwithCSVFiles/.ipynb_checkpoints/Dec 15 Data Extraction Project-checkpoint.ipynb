```python
import pandas as pd
import numpy as np

URL="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
dflist=pd.read_html(URL)

gdpdf=dflist[3] # extracted the df table i want
gdpdf.columns=range(gdpdf.shape[1]) # replaced the coloumn headings with numbers
# print(gdpdf)

filteredf=gdpdf.iloc[1:11,[0,2]] ## creating a new df which contains the top 10 economy and their imf based gdp
# print(filteredf)

filteredf.columns=['Country','GDP (Million USD)']
print(filteredf)

# print(type(filteredf[['GDP (Million USD)']]))

filteredf[['GDP (Million USD)']]=filteredf[['GDP (Million USD)']].astype(int) ## converting the gdp from str to int
filteredf[['GDP (Million USD)']]=filteredf[['GDP (Million USD)']]/1000 ## dividing the gdp by 1000
filteredf[['GDP (Million USD)']]=np.round(filteredf[['GDP (Million USD)']],2) ## rounding gdp to 2
filteredf.rename(columns={'GDP (Million USD)':'GDP (Billion USD)'})
print(filteredf)
```
