```python
import pandas as pd
import numpy as np

matrixdf=pd.DataFrame(np.array([[1,3,4],[5,6,7],[3,5,6]]),columns=['a','b','c'])
print(matrixdf)

# start going to learn about transform function which returns a self produced transformed dataframe according to the function specified inside it

matrixdf=matrixdf.transform(func= lambda x: x+10)
print(matrixdf)

# now going to find the sqrt of all those elements

matrixdf=matrixdf.transform(func=['sqrt'])
print(matrixdf)
```

```python
# Serializtion is the process of converting objects into a format suitable for transmitting files over the network or storing in files or in a database
# json library of the python uses dumb function to convert python object to json format
# json is a lightweight data interchange format which is independent of programming languages

import json

person = {
    'first_name' : 'Mark',
    'last_name' : 'abc',
    'age' : 27,
    'address': {
        "streetAddress": "21 2nd Street",
        "city": "New York",
        "state": "NY",
        "postalCode": "10021-3100"
    }
}

with open('samplejson.json','w+') as jsonobj:
    json.dump(person,jsonobj)

# there is dumps function which can be used for conversion only anytime without having to necessarily open the file

personjs=json.dumps(person,indent=4)

with open('samplejson2.json','w+') as jsonobj2:
    jsonobj2.write(personjs)

# now our python objects are serialized into json, we can revert them back into python obj so we can work with them data using the load() function

with open('samplejson2.json','r') as jsonobj3:
    despyobj=json.load(jsonobj3)

# print(type(despyobj))

# going to read a file using pd.read_json which creates a Dataframe

df2=pd.read_json('samplejson2.json')
print(df2)

```

```python
## We are going to Work with XML files by creating and parsing them through the help of ElementTree python module

import xml.etree.ElementTree as et

# going to create an xml element which will be later wrote to xml doc

mainelement=et.Element('person')
subelement=et.SubElement(mainelement,'details')
nameelement=et.SubElement(subelement,'name')
ageelement=et.SubElement(subelement,'age')
cityelement=et.SubElement(subelement,'city')

nameelement.text='messi'
ageelement.text='38'
cityelement.text='barcelona'

# going to transform the created element to Element treee

finalelementree=et.ElementTree(mainelement)

# going to create a file and write into my newly created element tree

with open('myxmlfile.xml','wb+') as myfile:
    finalelementree.write(myfile)

# now going to read from a xml file

parsedxml=et.parse('samplexml.xml')
rootoffile=parsedxml.getroot()

xmldatalist=[]

for node in rootoffile:
    name=node.find('name').text
    phonenumber=node.find('phonenumber').text
    birthday=node.find('birthday').text

    xmldatalist.append({'Name':name,'Phone Number':phonenumber,'Birthday':birthday})


xmldf=pd.DataFrame(xmldatalist)
# print(xmldf)

# there is also a pd.readxml func which can read xml file and convent them to df
# in the xpath we need to mention the parent nodes we want to migrate
df2=pd.read_xml('samplexml.xml',xpath='/contacts_list/person')
print(df2)

# going to save them to another file
df2.to_csv('newfile.csv',index=False)


    
    
    
```
