```python
# Data Analysis

# We are going to perform data analysis on Diabetes Dataset

# First we are going to download the Diabetes Dataset 

import requests
import pandas as pd

filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%205/data/diabetes.csv"

def download(url, filename):
    response = requests.get(url) ## when requesting files like csv, image, txt etc. the content will be retured as bytes stored in response.content attribute
    
    with open(filename, "wb") as f:
        f.write(response.content)

download(filename, "diabetes.csv")
df = pd.read_csv("diabetes.csv")

### download of the Dataset is Complete

# Statistical Overview of Dataset
# print(df.head()) ## print first 5 rows of the df
# print(df.shape) ## prints the dimension of the df like the rows and coloumns
# print(df.info())
# print(df.describe())

# Identifying and Handling Missing values
## we will be using python builtin function like isnull and notnull to identify them

### going to count the number of missing values in each coloumn

missing_datadf=df.isnull() # identifying the missing value df

for column in missing_datadf.columns.values.tolist():
    print(missing_datadf[column].value_counts())
    print(' ')

### checking if all datas are in the correct format
print(df.dtypes)

# Visualiztion is one of the most important ways of getting insights from the datasets
## seaborn and matplotlib are 2 of the most powerful python visualization libraries

### going to do piechart visulization
import matplotlib.pyplot as plt

labels=['Not Diabetic','Diabetic']
plt.pie(df['Outcome'].value_counts(),labels=labels,autopct='%0.02f%%')
plt.legend()
plt.show()
```
