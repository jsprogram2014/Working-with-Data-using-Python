```python
# Data Analysis

# We are going to perform data analysis on Diabetes Dataset

# First we are going to download the Diabetes Dataset 

import requests
import pandas as pd

filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%205/data/diabetes.csv"

def download(url, filename):
    response = requests.get(url) ## when requesting files like csv, image, txt etc. the content will be retured as bytes stored in response.content attribute
    
    with open(filename, "wb") as f:
        f.write(response.content)

download(filename, "diabetes.csv")
df = pd.read_csv("diabetes.csv")
print(df.head()) ## print first 5 rows of the df
print(df.shape) ## prints the dimension of the df like the rows and coloumns
### download of the Dataset is Complete

```
