```python
## we are going to create a Table/Dataframe of random Users which contains only Specific Coloumns
from randomuser import RandomUser
import pandas as pd

user_list=[]

def get_users():
    for user in RandomUser.generate_users(10):
        user_list.append({'Name':user.get_full_name(), 'Gender':user.get_gender(), 'City':user.get_city(), 'State':user.get_state(), 'Email':user.get_email(), 'DOB':user.get_dob(), 'Picture':user.get_picture()})
    return pd.DataFrame(user_list)
    # return user_list


df1=pd.DataFrame(get_users())
df1.to_csv('RandomUser1.csv')

## we have now created the dataframe containing only the details that we want of the users

```
